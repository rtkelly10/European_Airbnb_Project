### Step 6 - Create categorical plots

```python
# Histogram to see distribution
sns.histplot(df_sub['Total Price'], bins = 100, kde = True)
```

```python
# Apply a log transformation to the data
df_sub['Total Price Log'] = np.log1p(df_sub['Total Price'])
```

```python
# Plot the histogram with the transformed data
sns.histplot(df_sub['Total Price Log'], bins=10, kde=True)
plt.xlabel('Log-transformed Total Price')
plt.show()
```

```python
# Creating low score category for categorical plot
df_sub.loc[df_sub['Total Price Log'] < 5 , 'Price category'] = 'Low price'
```

```python
# Creating medium score category for categorical plot
df_sub.loc[(df_sub['Total Price Log'] >= 5) & (df_sub['Total Price Log'] < 6), 'Price category'] = 'Middle price'
```

```python
# Creating low score category for categorical plot
df_sub.loc[df_sub['Total Price Log'] >= 6 , 'Price category'] = 'High price'
```

```python
df_sub['Price category'].value_counts(dropna = False)
```

```python
# Creating categorical plot with Normalized Attraction Indez
sns.set(style="ticks")
cat_plot_GDP = sns.catplot(x='Normalized Attraction Index', y='Total Price Log', hue="Price category", data=df_sub)
```

#### This analysis reveals a positive correlation between property prices, represented by the logarithmically transformed 'Total Price,' and the 'Normalized Attraction Index.' As the attraction index increases, indicating higher proximity to attractions, there is a corresponding upward trend in property prices. The Seaborn categorical plot effectively captures this positive correlation by visually showcasing how the 'Total Price' varies across different levels of the 'Normalized Attraction Index,' offering valuable insights into the relationship between these two variables.
